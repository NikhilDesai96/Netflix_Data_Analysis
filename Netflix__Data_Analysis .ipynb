## Data Cleaning

```python
df.isnull().sum()
```

```python
df.drop(columns=["Unnamed: 12","Unnamed: 13","Unnamed: 14","Unnamed: 15","Unnamed: 16","Unnamed: 17","Unnamed: 18","Unnamed: 19","Unnamed: 20","Unnamed: 21","Unnamed: 22","Unnamed: 23","Unnamed: 24","Unnamed: 25"],inplace=True)
```

```python
## Handling missing values by replacing with 'Unknown'

df['director']=df['director'].fillna('Unknown')
df['cast']=df['cast'].fillna('Unknown')
df['country']=df['country'].fillna('Unknown')
```

```python
df.head()
```

```python
df.info()
```

```python
##  Dropping rows where important columns have missing values

df=df.dropna(subset=['date_added','rating','duration'])
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['date_added']=df['date_added'].str.strip()
```

```python
### Converting date_added column to datetime format

df['date_added']=pd.to_datetime(df['date_added'])
```

```python
## Extracting year from date_added

df['year_added']=df['date_added'].dt.year
```

```python
## Extracting month from date_added

df['month_added']=df['date_added'].dt.month
```

```python
df.head()
```

## Feature Engineering


```python
df['type'].unique()
```

```python
df['type'].value_counts()
```

```python
## Extracting main genre from listed_in column

df['main_genre']=df['listed_in'].str.split(',').str[0]
```

```python
df.head()
```

```python
## Extracting numeric duration from duration column

df['duration_int']=df['duration'].str.extract('(\d+)')
```

```python
df.head()
```

```python
## Converting extracted duration to numeric format

df['duration_int']=pd.to_numeric(df['duration_int'],errors='coerce')
```

## Exploratory Data Analysis (EDA)

```python
# Visualizing distribution of Movies vs TV Shows

df['type'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title("Movies vs Show")
plt.ylabel("")
plt.show()
```

Most content on Netflix consists of Movies compared to TV Shows.

```python
# Analyzing content added over the years

sns.countplot(x='year_added',data=df)
plt.xticks(rotation=90)
plt.title("Content Added Over Year")
plt.show()
```

```python
# Identifying top genres on Netflix

df['main_genre'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```

Drama is one of the most common genres available on Netflix.

```python
# Analyzing movie duration distribution

sns.histplot(df[df['type']=='Movie']['duration_int'], bins=20)
plt.title("Movie Duration Distribution")
plt.xlabel("Duration(mintues)")
plt.ylabel("Count")
plt.show()
```

Most movies on Netflix have a duration between 90 and 120 minutes.

## Additional Analysis (Country-wise)

```python
# Analyzing top countries producing Netflix content

movies=df[(df['type'])=='Movie']
movies = movies[movies.country !='Unknown']
top_country_movies = movies.country.value_counts().nlargest(5).sort_values(ascending=False)

top_country_movies.plot(kind='bar')
plt.title("Top 5 Countries Producing Movies")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

United States produces the highest number of movies on Netflix, followed by India.

```python
# Top Countries Producing TV Shows
tvshows=df[(df['type'])=='TV Show']
tvshows = tvshows[tvshows.country!='Unknown']
top_country_tvshows= tvshows.country.value_counts().nlargest(5).sort_values(ascending=False)

top_country_tvshows.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('TV Show')
plt.xticks(rotation= 45)
plt.show()
```

United States leads in TV show production on Netflix, followed by the United Kingdom and Japan.

```python
## Top Countries (Overall Content)
products=df[df.country!='Unknown']
top_movies_show= products.country.value_counts().nlargest(5).sort_values(ascending = False)

top_movies_show.plot(kind='bar')
plt.title("Top 5 Countries(Overall Content)")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

United States dominates overall Netflix content, followed by India and the United Kingdom.

## Final Insights
- Most content available on Netflix consists of Movies compared to TV Shows, indicating a stronger focus on film-based content.
- Content added on Netflix increased significantly after 2015, showing rapid platform growth and expansion.
- Drama is one of the most common genres on Netflix, highlighting audience preference for storytelling content.
- Most movies have a duration between 90 and 120 minutes, which reflects standard industry movie lengths.
- The United States produces the highest amount of content on Netflix, followed by India and the United Kingdom.
- India is a major contributor, especially in movie production, while countries like Japan contribute more to TV shows.
